# Thesis Repository Guide

This notebook summarizes the current active thesis workflow and folder layout.

## Repository Structure

- `simulation`: Gazebo world files, models, and RViz config
- `dataset`: bags, logs, and exported dataset artifacts
- `communication`: local communication-aware relay configuration
- `cooperative_sim`: active cooperative rule-based simulation stack
- `model_training`: export, training, notebooks, and comparison outputs
- `model_eval`: live learned-model evaluation

Local thesis-writing folders such as `thesis_template_hh` and `thesis_refrences` may exist in the workspace, but they are intentionally ignored and are not part of the tracked repository workflow.

```python
from pathlib import Path

THESIS_ROOT = Path('/home/basudeo/Documents/Thesis')
for folder in [
    'simulation',
    'dataset',
    'communication',
    'cooperative_sim',
    'model_training',
    'model_eval',
]:
    path = THESIS_ROOT / folder
    print(f'{folder}: {path} ->', 'exists' if path.exists() else 'missing')
```

## Active Pipeline

1. Run `cooperative_sim/scripts/run_sim_model.py` to generate cooperative rule-based runs.
2. Save bags and logs into `dataset`.
3. Export bags to episode frames with the `model_training/scripts` tools.
4. Build the shared sample table and train or compare models in the `model_training` notebooks.
5. Run live learned-model evaluation in `model_eval`.
6. Optionally enable `WifiRelay` or `BluetoothRelay` from `communication/basic/onmetpp/omnetpp.ini` for communication-aware live runs.

The active notebooks are scoped by purpose: `10_cv_baseline` for the non-neural reference, `20_cnn_lstm` for CNN--LSTM, `40_cnn_gnn_lstm` for CNN--GNN--LSTM, `50_cnn_gnn_transformer` for CNN--GNN--Transformer, and `60_cnn_gnn_lstm_transformer` for the combined CNN--GNN--LSTM--Transformer model.
