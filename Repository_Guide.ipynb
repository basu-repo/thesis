## Step-By-Step Installation

This section gives the minimum installation required to run the repository.

### Step 1. Operating system

Recommended:
- Ubuntu Linux

This repository is currently built and organized for a Linux environment.

### Step 2. Install ROS 2

Minimum:
- ROS 2 Humble

ROS 2 is required for:
- simulation nodes
- rule-based controllers
- rosbag recording
- controller communication

After installation, make sure this works:

```bash
source /opt/ros/humble/setup.bash
```

### Step 3. Install Gazebo and ROS-Gazebo bridge support

Required for:
- loading the world
- spawning Husky and UAV models
- publishing simulation topics into ROS
- RViz visualization

Minimum:
- Gazebo
- `ros_gz`
- RViz

### Step 4. Install Python 3.10 and create an environment

Recommended:
- Python 3.10
- one dedicated environment for dataset export and model evaluation

Example:

```bash
python3 -m venv thesis_env
source thesis_env/bin/activate
```

### Step 5. Install Python packages

Minimum package set:

```bash
pip install numpy pandas matplotlib torch scikit-learn jupyter rosbags
```

Why these are needed:
- `numpy`: arrays and `.npy` assets
- `pandas`: result summaries
- `matplotlib`: plots
- `torch`: models
- `scikit-learn`: ROC, PR, AUC, classification metrics
- `jupyter`: notebooks
- `rosbags`: exporting rosbag data to extracted dataset format

### Step 6. Optional communication installation

Only needed for communication-aware experiments.

Minimum:
- OMNeT++

This is not required for the base pipeline of:
- rule-based simulation
- dataset export
- notebook training and evaluation

## Pipeline Explanation

This repository is intended to be used in the following order.

### Stage 1. Simulation world

Folder:
- `01_simulation_world`

What happens here:
- Gazebo world is defined
- robot models are stored
- RViz setup is kept

Use this stage when changing:
- world layout
- model assets
- simulation visualization

### Stage 2. Rule-based simulation

Folder:
- `02_rule_based`

What happens here:
- Husky and UAV are controlled with rule-based logic
- rosbag data is recorded
- maneuver-rich behavior is generated for training

### Stage 3. Dataset export

Folder:
- `03_dataset`

What happens here:
- bags are exported into extracted dataset format
- `frames.jsonl` is written
- `.npy` assets are saved
- extracted hybrid maneuver dataset becomes ready for notebooks

Preferred extracted output:
- `/media/basudeo/1044063744061FD8/hybrid_maneuver_dataset`

Fallback extracted output:
- `03_dataset/hybrid_maneuver_dataset`

### Stage 4. Model evaluation

Folder:
- `04_model_evaluation`

What happens here:
- extracted dataset is loaded
- split is built or reused
- model notebooks are run one at a time
- metrics, weights, plots, and comparison summaries are saved

Saved outputs go to:
- `04_model_evaluation/results`
- `04_model_evaluation/model_weights`
- `04_model_evaluation/plots`
- `04_model_evaluation/comparison_exports`
- `04_model_evaluation/splits`

### Stage 5. Final learned-model simulation

Folder:
- `05_final_eval_sim`

What happens here:
- the best model is selected
- that model is loaded back into simulation
- learned behavior is tested in the Gazebo environment

### Stage 6. Communication layer

Folder:
- `06_Communication`

What happens here:
- communication-aware testing can be added
- OMNeT++ can be enabled when required
- external team work remains separated under `external_team`

This stage is optional.

## Practical Run Order

If you are starting fresh, use this order:

1. Verify ROS 2, Gazebo, and Python environment
2. Check `02_rule_based/scripts/project_paths.py`
3. Check external dataset path availability
4. Run the rule-based simulation
5. Record bag data
6. Export the dataset
7. Open `04_model_evaluation/notebooks`
8. Restart the kernel before each notebook
9. Run notebooks one by one
10. Compare models
11. Select the best model
12. Use the selected model in `05_final_eval_sim`
13. Add communication only if required
