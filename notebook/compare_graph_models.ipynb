# Compare CV, GNN-LSTM, and CNN-GNN-LSTM

This notebook reads the saved summary JSON files from `models/` and compares the offline results for Constant Velocity, GNN-LSTM, and CNN-GNN-LSTM side by side.


```python
from pathlib import Path
import json

MODELS_DIR = Path.home() / 'Documents/Thesis/models'
GNN_SUMMARY = MODELS_DIR / 'summary_gnn_graph_done.json'
CNN_GNN_SUMMARY = MODELS_DIR / 'summary_cnn_gnn_graph_done.json'

def load_summary(path: Path):
    if not path.exists():
        return None
    return json.loads(path.read_text())

gnn = load_summary(GNN_SUMMARY)
cnn_gnn = load_summary(CNN_GNN_SUMMARY)

gnn, cnn_gnn is not None
```

## Build A Comparison Table

The table below uses the `comparison` blocks from the saved summary files. If one summary is missing, the corresponding row will be marked as missing.


```python
cv_metrics = None
gnn_metrics = None
cnn_gnn_metrics = None

if gnn is not None:
    cv_metrics = gnn.get('comparison', {}).get('constant_velocity')
    gnn_metrics = gnn.get('comparison', {}).get('gnn_lstm')

if cnn_gnn is not None:
    if cv_metrics is None:
        cv_metrics = cnn_gnn.get('comparison', {}).get('constant_velocity')
    cnn_gnn_metrics = cnn_gnn.get('comparison', {}).get('cnn_gnn_lstm')

rows = [
    ('Constant Velocity', cv_metrics),
    ('GNN-LSTM', gnn_metrics),
    ('CNN-GNN-LSTM', cnn_gnn_metrics),
]

for name, metrics in rows:
    if metrics is None:
        print(f'{name:<18} missing')
    else:
        print(
            f"{name:<18} "
            f"ADE={metrics['ADE']:.6f} "
            f"FDE={metrics['FDE']:.6f} "
            f"RMSE={metrics['RMSE']:.6f}"
        )
```
